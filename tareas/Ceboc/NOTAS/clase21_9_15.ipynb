# Series de Taylor

Van a ser serias de Taylor no en el sentido de calculo uno, sino en el sentido de diferenciación automática. NO van a ser las series de Taylor a las que estamos acostumbrados, van a ser series de Taylor cuyos coeficientes serán las entradas de algún "n-ual". Es claro que necesitamos alguna manera de eficientar la programación.
$$
f(x)=f_0 + f_1 (x-x_0)+f_2 (x-x_0)^2 + \cdots + f_k (k-x_0)^k
$$
donde 
$$
f_k = \frac{f^{(k)}(x_0)}{n!}
$$

si $f\in C_\infty$ la serie anterior tiene infinitos coeficientes. Si conocemos los coeficientes $f_k$, podemos aproximar a la función

$$
f(x)\sim(f_0,f_1,f_2,\dots,f_k,\dots)
$$

digamos ¿Qué sería? (se sobreentiende la evaluación en $x_0$)
$$
f(x)\pm g(x)=(f_0\pm g_0)+(f_1\pm g_1)(x-x_0)+\cdots (f_k\pm g_k)(x-x_0)^k + \cdots
$$

$$
f(x)\cdot g(x)=f_0 + g_0 + (f_0 g_1 + f_1 g_0)(x-x_0) + (f_0 g_2 + f_1 g_1 + f_2 g_0)(x-x_0)^2+ \cdots + \sum\limits_{n=0}^k f_n \cdot g_{k-n} + \cdots
$$

$$
f(x)\div g(x) = \frac{1}{g_0}(f_k - \sum\limits_{i=0}^{k-1} (f\div g)_i \cdot g_{k-i}
$$

esta última relación no es tan obvia

$$
f(x) \div g(x) = \frac{\sum\limits_{k=0}^\infty f_k \cdot (x-x_0)^k}{\sum\limits_{k=0}^{\infty} g_k \cdot (x-x_0)^k} = \sum\limits_{k=0}^{\infty} (f \div g)_k \cdot (x-x_0)^k
$$

despejando a $f_k$

$$
\sum\limits_{k=0}^\infty f_k \cdot (x-x_o)^k = \sum\limits_{k=0}^\infty (f \div g)_i \cdot (x-x_0)^k \cdot \sum\limits_{k=0}^{\infty} g_k \cdot (x-x_0)^k
$$

ahora 
$$
f_k = \sum\limits_{i=0}^k (f\div g)_i  \cdot g_{k-i} = \sum\limits_{i=0}^{k-1}(f \div g)_i \cdot g_{k-i} + (f \div g)_k \cdot g_0
$$

$$
\frac{1}{g_0} \big{(}f_k - \sum\limits_{i=0}^{k-1} (f+g)_i \cdot g_{k-i} \big{)}= (f \div g)_k
$$

Hay que tener cuidado con que $g_0$, en caso de serlo hay que ver lo que da en la penúltima ecuación, el último término que sacamos de la suma, se hace sero, con lo que hay que sacar el que sigue y despejarlo.

Como podemos acercar la función tanto como querramos, pero la computadora no puede hacer esto, hay que truncar la serie, de hecho, hasta lo más lejos que nos podemos ir en la serie es en donde tengamos un error del orden de la computadora, irnos más lejos no tiene sentido hacerlo. Si $f \notin C_n$ tal que $n$ sea lo suficientemente grande no tendría siquiera sentido hacer la serie de Taylor de orden $n$, pues los coeficientes de la serie irían divergiendo.

Vamos a hacer lo siguiente

$$
e^{g(x)} = \sum\limits_{k=0}^\infty (e^g)_k \cdot (x-x_0)^k
$$

recordemos que 

$$
\frac{d}{dx} e^{g(x)}=g'(x) \cdot e^{g(x)}
$$

con esto veamos cómo podemos reescribir la suma

$$
\sum\limits_{k=1}^\infty k (e^g)_k (x-x_0)^k-1=\sum\limits_{k=1}^\infty k g_k (x-x_0)^{k-1} \cdot \sum\limits_{k=0}^\infty (e^g)_k (x-x_0)^k
$$


$$
\sum\limits_{k=1}^\infty k (e^g)_k (x-x_0)^k= \sum\limits_{k=0}^\infty k g_k(x-x_0)^k \cdot \sum\limits_{k=0}^\infty (e^g)_k (x-x_0)^k
$$

la exponencial de la serie de Taylor de una función con diferenciación automática es:

$$
k(e^g)_k = \sum\limits_{i=1}^k i g_i (e^g)_{k-i}
$$

en el camino resolvimos la ecuación diferencial, en general esa es la idea para resolver diferenciales, salvo que no siempre es tan fácil.
$$
(e^g)_k = \begin{cases} e^{g_0} &\mbox{si } k = 0 \\
\frac{1}{k}\sum\limits_{i=1}^k i g_i (e^g)_{k-i} & \mbox{si } k > 0. \end{cases}
$$

ahora a otra cosa
$$
\frac{d}{dx} \ln{g(x)} = \frac{g'(x)}{g(x)}
$$

aplicando la serie de Taylor

$$
\ln(g(x)) = \sum\limits_{i=0}^\infty (\ln{g})_k (x-x0)^k
$$

y ahora sacamos la serie de Taylor de la derivada

$$
\sum\limits_{k=1}^\infty k (\ln{g})_k (x-x_0)^{k-1} = \frac{\sum\limits_{k=1}^\infty k g_k (x-x_0)^{k-1}}{\sum\limits_{k=1}^\infty  g_k (x-x_0)^{k}}
$$

con esto despejamos

$$
\sum\limits_{k=1}^\infty k (\ln{g})_k (x-x_0)^{k-1} \cdot \sum\limits_{k=1}^\infty  g_k (x-x_0)^{k} = \sum\limits_{k=1}^\infty k g_k (x-x_0)^{k-1}
$$

así

$$
k g_k = \sum\limits_{i=1}^k i (\ln{g})_i g_{k-i} = \sum\limits_{i=1}^{k-1} i (\ln{g})i g_{k-i} + k(\ln{g})_k g_0
$$

y así como lo habíamos hecho en el caso anterior

$$
\frac{1}{g_0}\big{(} g_k - \frac{1}{k} \sum\limits_{i=1}^{k-1} i (\ln{g})_i g_{k-i}  \big{)} = (\ln{g})_k
$$


$$
(\ln{g})_k = \begin{cases} \ln{g_0} &\mbox{si } k = 0 \\
\frac{1}{g_0}\big{(} g_k - \frac{1}{k} \sum\limits_{i=1}^{k-1} i (\ln{g})_i g_{k-i}  \big{)} & \mbox{si } k > 0. \end{cases}
$$

entonces con esto ya podemos exponenciar

$$
e^{g(x) \ln{f(x)}}=f(x)^{g(x)}
