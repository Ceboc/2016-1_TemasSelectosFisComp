Ahora un último ejemplo de Taylor. Un resumen de lo que se obtuvo ayer.

Se había visto que íbamos a representar a las funciones por sus coeficientes de Taylor.

$$
f \sim (f_0, f_1, f_2, \dots , f_k ,\dots)
$$

$$
f(x)=f_0 + f_1(x-x_0) +f_2(x-x_0)^2 + \cdots + f_k(x-x_0)^k
$$

donde

$$
f_k = \frac{f^{(k)}(x_0)}{k!}
$$

y recordemos además lo que se había definido ayer

$$
f(x)\pm g(x)= (f_0 \pm g_0)+ (f_1 \pm f_2)(x+x_0) + \cdots
$$

para el producto vimos que el coeficiente $k$-ésimo era

$$
(f(x)g(x))_k =\sum\limits_{i=0}^k f_i g_{k-i}
$$

además habíamos sacado que 

$$
e^{f(x)}\Rightarrow (e^{f(x)})_k = \begin{cases} e^{f_0}&\mbox{si } k=0 \\
\frac{1}{g_0}\big{(} \cdots & \mbox{si } k>0. \end{cases}
$$

y recordemos que en el camino se resolvió

$$
\frac{d}{dx} e^{f(x)}=f'e^{f(x)}
$$

$$
f(x)^{g(x)}=e^{g(x) \ln{f(x)}}
$$

la serie de Taylor de una función $f(x)$

$$
f(x) = \sum\limits_{n=0}^k f_0 (x-x_0)^n + F(x-x_0)^k+1
$$

¿Con lo que tenemos podemos resolver?

$$
\dot{y}=f(y)
$$

recordemos que su solución es toda una familia de soluciones, realmente necesitamos una condición inicial.

$$
y_0 = f(t_0)
$$

hagamos lo siguiente

$$
\sum\limits_{t=0}^\infty k y_k (t-t_0)^{k-1}=\sum\limits_{k=0}^\infty f_k (t-t_0)^k
$$

donde el primer término lo podemos ver como
$$
\sum\limits_{k=0}^\infty (k+1)y_{k+1}(t-t_0)^k
$$
y podemos despejar 
$$
y_{k+1}=\frac{f_{k+1}}{k+1}
$$

El problema es que la computadora no sabe hasta donde parar $k$, además necesitamos integrar en el tiempo, ie. para obtener la evolución de la dinámica del sistema requerimos calcular nuevamente para $t_1, t_2, t_3, \dots, t_n,\dots$ por lo que necesitamos conocer los nuevos coeficientes de Taylor para cada tiempo, nada nos asegura que los coeficientes se van a quedar igual, es decir, luego de hacer la dinámica al primer paso, necesitamos conocer los nuevos coeficientes para la nueva condición inicial que será el último valor de la función calculado, a cada paso, necesitamos calcular las condiciones inciciales. No se puede hacer con los mismos coeficientes pues sería equivalente a considerar siempre las mismas condicioenes iniciales con lo que jamás conoceríamos las dinámica del sistema. Lo que procede es evolucionar al sistema, ¿Cómo escojemos la $\Delta t$ ?. Tenemos que reevaluar cada vez, por que es mejor la aproximación en rededor de cada tiempo. Claro la serie de Taylor converge a la función pero, no es nuestro caso, es truncada, pero principalmente por que los coeficientes cambian a cada punto, a cada punto la serie de Taylor para la función es distinta, aunque como se vio en Cálculo esta converga *ad infinitum* a la función.

Necesitamos conocer una $\epsilon$  tal que nos funcione para un $\tau$  que nos interese. Sea $\delta = t-t_0$ y $P$ el coeficiente al que cortamos la serie.

$$
|y_p| (t-t_0)^P<\epsilon
$$

$$
\delta < \big{|}\frac{\epsilon}{|y_P|}\big{|}^{\frac{1}{P}}
$$

Esto nos asegura que usemos una aproximación tan precisa que sea innecesaria, no queremos trabajar debajo de la $\epsilon$ de la máquina, no tiene utilidad hacer esto.
