EL chiste ahora es hacer operaciones con la presición de la máquina, con su $\epsilon$. 
Definimos el vector dual:
$$
\vec{u}=(u_o,u_o')
$$
donde
$$
u_o=u(x_o)
$$
y
$$
u'_o=u'(x_o)=\frac{d}{dx}u(x_o)
$$

Recordemos de variable compleja que nos definieron los númeors complejos como $a+ib$, podemos usar esta definición de vector para los números complejos pero vamos a ir a otro lugar, esto sólo fue un paréntesis cultural.

Las estructuras tipo que se vieron ayer sirven para construir estos bichos. 
Propiedades:
$$
\vec{u}+\vec{w}=(u_o\pm w_o, u_o'+w_o')
$$

$$
\vec{u} \cdot \vec{w} = (u_o \cdot w_o,u_o \cdot w_o'+ u_o' \cdot w_o')
$$

$$
\frac{\vec{u}}{\vec{w}}=(\frac{u_o}{w_o},\frac{w_o u_o' - u_o' w_o}{w_o^2})=(\frac{u_o}{w_o}, \frac{u_o' - (\frac{u_o}{w_o})w_o'}{w_o})
$$
Son estas las mejores opciones para realizar las operaciones, en particular la última propiedad, la segund igualdad es más eficiente por que usamos algo que previamente ya habíamos calculado.

La última

$$
\vec{u}^\alpha = (u^\alpha, \alpha u_o^{\alpha-1}u_o')
$$

con esto podemos pensar en lo que sigue
$$
\vec{c}=(c,0),\space c\in\mathbb{R}
$$

$$
\vec{x_o}=(x_o,1),\space x_o \in \mathbb{R}
$$

si 
$$
f(x)=3x^2-4x+5
$$
podemos reescribirlo como
$$
f(x)=\vec{3}\vec{x}^2-\vec{4}\vec{x}+\vec{5} \\
 = (3,0)\cdot (1,1)^2 - (4,0)\cdot(1,1)+(5,0) \\
 =(3,0)\cdot (1^2,2\cdot 1\dot 1)-(4,0)\cdot(1,1)+(5,0)
 =(3-4+5,6-4)=(4,2)
$$

¡La segunda entrada es la derivada! La chamba ahora es enseñarle a julia a calcular, tal que la segunda entrada del vector dual es la derivada, es más el valor de esta derivada es exacta, el error que contiene es el error de la computadora. A priori todos la fuciones racionales las podemos a priori derivar.
